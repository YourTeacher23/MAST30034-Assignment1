```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from random import gauss
import math
from scipy.stats import pearsonr

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
NUM_SOURCE = 6
X1 = 21
X2 = 21
V = 441
N = 240
LABEL = ['First', 'Second', 'Third', 'Fourth', 'Fifth', 'Sixth']
```

```python
Apcr = pd.read_csv("../data/apcr.csv")
Dpcr = pd.read_csv("../data/dpcr.csv")

Apcr = Apcr.reset_index(drop=True)
Dpcr = Dpcr.rename(columns = {'V1':'0', 'V2':'1', 'V3':'2', 'V4':'3', 'V5':'4', 'V6':'5'})

A_pcr = []
apcr = np.array(Apcr)

for row in range(NUM_SOURCE):
    A_pcr.append(apcr[row].reshape((X1, X2)))
```

```python
for i in range(NUM_SOURCE):
    fig, axes = plt.subplots(1, 2, figsize=(10,3))
    ax1 = sns.lineplot(data= Dpcr[str(i)], ax = axes[0])
    ax1.set(xlabel=LABEL[i]+' Dpcr')
    ax2 =sns.heatmap(data = A_pcr[i], ax = axes[1])
    ax2.set(xlabel=LABEL[i]+' Apcr')
    plt.tight_layout()
    plt.savefig('../plots/Q2_5-' + str(i+1) + '.png')
    plt.show()
```
